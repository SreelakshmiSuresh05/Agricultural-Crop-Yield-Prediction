Outlier Detection and Removal:Use box plots ,interquartile range(IQR) to detect outliers


```python
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_cols=['Rainfall_mm','Temperature_Celsius','Days_to_Harvest','Yield_tons_per_hectare']
for i in numeric_cols:
  plt.figure(figsize=(4,2))
  sns.boxplot(data=df,y=i)
  plt.show()
```

```python
numeric_cols=['Rainfall_mm','Temperature_Celsius','Days_to_Harvest','Yield_tons_per_hectare']
for col in numeric_cols:
  Q1= df[col].quantile(0.25)
  Q3= df[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound = Q3+1.5*IQR
  df=df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]
print(df)
```

To convert object type and boolean data into a format suitable for machine learning models,we're using an encoding method called GetDummies. This process transforms each category into a new binary column..



```python
df1=pd.get_dummies(df[['Region','Soil_Type','Crop','Irrigation_Used','Fertilizer_Used','Weather_Condition']],drop_first='True',dtype=int)
df1
```

```python
new_df=pd.concat([df,df1],axis=1)
new_df
```

```python
new_df.drop(['Region','Soil_Type','Crop','Irrigation_Used','Fertilizer_Used','Weather_Condition'],axis=1,inplace=True)
```

```python
x=new_df.drop(columns=['Yield_tons_per_hectare'])
x
```

```python
y=new_df['Yield_tons_per_hectare']
y
```

Splitting the data into training and testing sets(70% train,30% test)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Standardizing Features for Linear Regression

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
Scaler.fit(x_train)
x_train=Scaler.transform(x_train)
x_test=Scaler.transform(x_test)
```

**Model Implimentations and Evaluations**

Model creation(Linear Regression)

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_pred_lr=lr_model.predict(x_test)
```

Evaluation Metrics(Linear Regression)

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
print("R2 SCORE                            :",r2_score(y_test,y_pred_lr))
print("MEAN ABSOLUTE ERROR            (MAE):",mean_absolute_error(y_test,y_pred_lr))
print("MEAN ABSOLUTE PERCENTGAE ERROR(MAPE):",mean_absolute_percentage_error(y_test,y_pred_lr))
k=mean_squared_error(y_test,y_pred_lr)
print("MEAN SQUARED ERROR             (MSE):",k)
print("ROOT MEAN SQUARED ERROR       (RMSE):",np.sqrt(k))
```

Model creation(Decision Tree Regressor)

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
dt_model.fit(x_train,y_train)
y_pred_dt=dt_model.predict(x_test)
```

Evaluation Metrics(Decision Tree Regressor)

```python
print('R2 SCORE                            :',r2_score(y_test,y_pred_dt))
print('MEAN ABSOLUTE ERROR (MAE)           :',mean_absolute_error(y_test,y_pred_dt))
print('MEAN ABSOLUTE PERCENTAGE ERROR(MAPE):',mean_absolute_percentage_error(y_test,y_pred_dt))
k=mean_squared_error(y_test,y_pred_dt)
print('MEAN SQUARED ERROR (MSE)            :',k)
print('ROOT MEAN SQUARED ERROR (RMSE)      :',np.sqrt(k))
```

Model creation (Random Forest Regressor)

```python
from sklearn.ensemble import RandomForestRegressor
rf_model= RandomForestRegressor(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
y_pred_rf=rf_model.predict(x_test)
```

Evaluation Metrics(Random Forest Regression)

```python
print("R2 SCORE                             :",r2_score(y_test,y_pred_rf))
print("MEAN ABSOLUTE ERROR (MAE)            :",mean_absolute_error(y_test,y_pred_rf))
print("MEAN ABSOLUTE PERCENTGAE ERROR (MAPE):",mean_absolute_percentage_error(y_test,y_pred_rf))
k=mean_squared_error(y_test,y_pred_rf)
print("MEAN SQUARED ERROR (MSE)             :",k)
print("ROOT MEAN SQUARED ERROR (RMSE)       :",np.sqrt(k))
```
